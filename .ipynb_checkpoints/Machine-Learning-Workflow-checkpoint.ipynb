```python
%matplotlib inline
import numpy as np
import pandas as pd
#import seaborn as sns
#from keras.layers import Dense
import matplotlib.pyplot as plt
#from keras.optimizers import SGD
#from keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
#from keras.utils.np_utils import to_categorical
from sklearn.model_selection import train_test_split

#X is a set of parameters defined at the point of entry for each trade.
#y is a list of returns for each trade.
X = pd.read_csv("Data/X.csv",header=None).values
y = pd.read_csv("Data/y.csv",header=None).values.reshape(-1,1)

#X = np.random.randn(8000,23)
#y = (X[:,1]*X[:,3] + 20*X[:,22]).reshape(8000,1)
print("X shape = {} :: y.shape = {}".format(X.shape,y.shape))

X_train, X_test, y_train, y_test = train_test_split(
        X, y,test_size = 0.5)
```

```python
from sklearn.cluster import KMeans
import ipywidgets
```

```python
n = 5
kmeans = KMeans(n_clusters=n)
kmeans.fit(X_train)
l_train = kmeans.predict(X_train)
l_test = kmeans.predict(X_test)
```

```python
avgprofit = [sum(y[:,:])/len(y[:,:])]
npoints = [len(y[:,:])]
xx = [-1]
l = l_test
yc = y_test
for j in range(0,n): 
    avgprofit.append(sum(yc[l==j,:])/len(yc[l==j]))
    npoints.append(len(yc[l==j,:]))
    xx.append(j)
```
