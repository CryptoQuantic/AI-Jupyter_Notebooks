```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from keras import layers
from keras.layers import Dense
import matplotlib.pyplot as plt
from keras.optimizers import SGD
from keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
from keras.utils.np_utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
#genearte a complex time series
def generateTimeSeries(length,p1,ncomps):
    xx = np.arange(length)
    y = 0;
    for j in range(0,ncomps):
        period = np.random.rand(1)*5
        phase  = np.random.randint(length)
        y =  y+np.sin((xx+phase)*period/6.28)/period
    return y

#takes a time series and splits it at a fraction of its length, and returns a point at the and for predicting.
def generateXy(series,fraction):
    xlen = series.shape[0]*fraction
    xlen = int(round(xlen))
    X = series[0:xlen]
    y = series[-1]
    return (X,y)
```

```python
for j in range(0,3000):
    if j == 0:
        A,y = generateXy(generateTimeSeries(100,1,13),.5)
        A = A.reshape(1,-1)
        y = y.reshape(1,-1)
    else:
        Xtemp,ytemp = generateXy(generateTimeSeries(100,1,13),.5)
        Xtemp = Xtemp.reshape(1,-1)
        ytemp = ytemp.reshape(1,-1)
        A = np.concatenate((A,Xtemp),axis = 0)
        y = np.concatenate((y,ytemp),axis = 0)
```

```python
xx = range(0,50)
_ = plt.plot(xx,A[18],xx,A[22])
```

```python
_ = plt.plot(y)
```

```python
model = Sequential()
model.add(Dense(50,input_shape=(50,),activation = 'relu'))
model.add(Dense(32,activation = 'relu'))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
model.summary()
```

```python
h = model.fit(A, y, batch_size=128, epochs=200, verbose=0)#,validation_data=(X_test, y_test_categorical))
plt.plot(h.history['loss'])
#plt.plot(h.history['val_loss'])
```

```python
yfit = model.predict(A)
```

```python
plt.plot(y,yfit,'bo')
```
